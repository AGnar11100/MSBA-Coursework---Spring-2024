## Example: Keeping Tabs on Congress

Note: this work is based on the example at Chapter 9 in the book of Data Science from Scratch

We are interested in what Congress is saying on the price change. If congree talking "Small Business"

The page of "https://www.house.gov/representatives" shows links to all the representatives' website. So let's start by collecting all of the URLs linked to from that page:

```python
from bs4 import BeautifulSoup
import requests

url = "https://www.house.gov/representatives"
text = requests.get(url).text
soup = BeautifulSoup(text, 'html.parser')
print(soup.prettify())
```

From the source code, we can see that all the links are between tag 'a' and have attribute "href". So, we find all the URLs fits this condition.

```python
all_urls = [a['href'] for a in soup('a')  if a.has_attr('href')]
all_urls
print(len(all_urls))
print(all_urls)
```

However, as we can see, this returns way too many URLs. If we look at the URLs we want to collect, we can see that they start with either *http://* or *https://*, have some kind of name, and end with either *.house.gov* or *.house.gov/*.

```python
import re
regex = r"^https?://.*\.house\.gov/?$"
# this requires the URL must start with http:// or https://
# must end with .house.gov or .house.gov/

# we can test some cases.
assert re.match(regex, "http://joel.house.gov")
assert not re.match(regex, "joel.house.gove")

# apply to our original list
good_urls = [url for url in all_urls if re.match(regex, url)]
print(len(good_urls))
print(good_urls)
```

There are only 435 representatives, so our list is still way too many. When we look at the list, we can see that there are a lot of duplicates. We can use set to get rid of them.

```python
good_urls = list(set(good_urls))
print(good_urls)
print(len(good_urls))
```

Just getting the link for representatives is not enough. Remember what we want to do is to find all the representatives who have press release about "chatgpt". So we need to find the URLS contains all the representatives press release. For example, we start at the first in our *good_urls* list.

```python
html = requests.get("https://tenney.house.gov/").text
soup = BeautifulSoup(html, 'html.parser')

print(soup)

```

From the source code, we can see that the link for the press release is stored at *a* tag with Press Releases as its content.

```python
links = {a['href'] for a in soup('a') if 'press releases' in a.text.lower()}

print(links)
```

Note that this link is a relative link, which means it should be add behind the originating link.

```python
from typing import Dict, Set

press_releases: Dict[str, Set[str]] = {}

for house_url in good_urls:
    html = requests.get(house_url).text
    soup = BeautifulSoup(html, 'html.parser')
    
    pr_links = {a['href'] for a in soup('a') if 'press releases' in a.text.lower()}
#    if pr_links starts with https
#    else:
    for pr_link in pr_links:
        url=f"{house_url}/{pr_link}"
        print(url)
 
    
    #print(f"{house_url}/{pr_links}")
    press_releases[house_url] = pr_links
```

```python
print(len(press_releases))
```

We use a function to check if a keyword is in the content of the link or not. 

```python
def paragraph_mentions(text:str, keyword: str) -> bool:
    """
    Returns True if a <p> inside the text mentions {keyword}
    """
    soup = BeautifulSoup(text, 'html.parser')
    paragraphs = [p.get_text() for p in soup('p')]
    
    return any(keyword.lower() in paragraph.lower() for paragraph in paragraphs)
```

We can run some cases to test this function.

```python
text ="""<body><h1>Facebook</h1><p>Twitter</p>"""
assert paragraph_mentions(text, "twitter")  #True
assert not paragraph_mentions(text, "facebook")
```

```python
for house_url, pr_links in press_releases.items():
    for pr_link in pr_links:
        url=f"{house_url}/{pr_link}"
        text = requests.get(url).text
               
        if paragraph_mentions(text, 'Small Business'):
            print(f"{house_url}")
            break
        
```
