# Question 1

Scraping a specific text in html

Given this book webpage,
https://www.mheducation.com/highered/product/process-parenting-brooks/M9780078024467.html
, 

can you please scrape the ISBN10 and ISBN13 code for the book using re lib? 

```python
from urllib.request import urlopen
import re

url = "https://www.mheducation.com/highered/product/process-parenting-brooks/M9780078024467.html"
htmlfile=urlopen(url)
text=str(htmlfile.read())
print(text)
```

```python
ISBN10=re.findall('<span class="isbn">(.*?)</span>',text)
print(ISBN10)
ISBN13=re.findall('<span class="isbn13">(.*?)</span>',text)
print(ISBN13)
```

# Question 2

Scraping a specific text in html

Continued with the previouse exercise, can you please scrape the ISBN10 and ISBN13 code for the book, using beautiful soup lib?

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as BS

url = "https://www.mheducation.com/highered/product/process-parenting-brooks/M9780078024467.html"
htmlfile=urlopen(url)
soup = BS(htmlfile,'html.parser')
#print(soup)
ISBN10=soup.find_all('span',attrs={"class":"isbn"})
ISBN13=soup.find_all('span',attrs={"class":"isbn13"})
print(ISBN10, ISBN13)
```
